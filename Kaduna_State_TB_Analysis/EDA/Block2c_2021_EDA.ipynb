```python
block2c_2021_flattened = pd.melt(data,id_vars=['Year','Quarter','LGA','Sex'],
                        value_vars=['0-4','5-14','15-24','25-34','35-44','45-54','55-64','>=65'],
                        var_name = 'Age_Gr',
                        value_name = 'TB_Cases')
```

```python
block2c_2021_flattened
```

```python
sns.lineplot(block2c_2021_flattened['TB_Cases'])
```

```python
sns.histplot(block2c_2021_flattened['TB_Cases'])
```

## Bivariate Analysis

LGA WISE TB CASES

```python
LGAwise_TB_Cases = block2c_2021_flattened.groupby('LGA')['TB_Cases'].sum().reset_index()
```

```python
fig = plt.figure(figsize=(40,10))

sns.barplot(
    data = LGAwise_TB_Cases,
    x = 'LGA',
    y = 'TB_Cases'
)

plt.title('LGA-wise TB Cases',fontsize= 30,fontweight='bold')
plt.xlabel('LGA',fontsize= 20,fontweight='bold')
plt.ylabel('Total TB Cases',fontsize= 20,fontweight='bold')
plt.xticks(fontsize= 20,fontweight='bold',rotation=60)
plt.yticks(fontsize= 20,fontweight='bold')
```

Age wise Tb cases

```python
AgeGr_wise_TB_Cases = block2c_2021_flattened.groupby('Age_Gr')['TB_Cases'].sum().reset_index()
```

```python
plt.figure(figsize = (10,10))

sns.barplot(
    data = AgeGr_wise_TB_Cases,
    x = 'TB_Cases',
    y = 'Age_Gr',
    orient = "h",
    order = ['0-4','5-14','15-24','25-34','35-44','45-54','55-64','>=65']
)

plt.title('Age_Gr wise TB Cases')
plt.xlabel('TB Cases')
plt.ylabel('Age Gr')
```

```python
#category wise distribution
category_count =block2c_2021_flattened['LGA'].value_counts().reset_index()
category_count.columns = ['LGA', 'TB_Cases']

# Set the style of the plot (optional)
sns.set(style="whitegrid")

# Create the count plot
plt.figure(figsize=(12, 6))
sns.barplot(data=block2c_2021_flattened, x='LGA', y='TB_Cases', palette="Set1")

# Customize the plot labels and appearance
plt.title('LGA Distribution')
plt.xlabel('LGA')
plt.ylabel('TB_Cases')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## Multivariate Analysis

```python
block2c_2021_flattened.head()
```

```python
Quarterwise_Trend_Over_LGA = block2c_2021_flattened.groupby(['LGA','Quarter'])['TB_Cases'].sum().reset_index()
```

```python
Quarterwise_Trend_Over_LGA
```

```python
LGAs = list(pd.unique(Quarterwise_Trend_Over_LGA['LGA']))
```

```python
fig,axes = plt.subplots(4,6,figsize=(50,30))

i=0
j=0

for lga in LGAs:
  sns.lineplot(
      ax = axes[i,j],
      data = Quarterwise_Trend_Over_LGA.loc[Quarterwise_Trend_Over_LGA['LGA']==lga],
      x = 'Quarter',
      y = 'TB_Cases',
      hue = 'LGA',
      marker = 'o',
      markersize=5
  )
  axes[i,j].set_title(lga, fontsize=20, fontweight='bold')
  axes[i,j].set_xticks([1,2,3,4],[1,2,3,4])
  axes[i,j].set_ylim(bottom=0)
  j = j+1
  if j==6:
    i = i+1
    j=0

plt.suptitle('Quarterwise trend of TB Cases In Each LGA', fontsize=35, fontweight='bold')
```

```python
Quarterwise_Trend_Over_Age = block2c_2021_flattened.groupby(['Quarter','Age_Gr'])['TB_Cases'].sum().reset_index()
```

```python
Quarterwise_Trend_Over_Age
```

```python
plt.figure(figsize = (10,10))

sns.lineplot(
    data = Quarterwise_Trend_Over_Age,
    x = 'Quarter',
    y = 'TB_Cases',
    hue = 'Age_Gr'
)

plt.title('Quarter-wise Trend Of TB Cases For Different Age Groups')
plt.xlabel('Quarter')
plt.ylabel('TB Cases')
plt.xticks([1,2,3,4],[1,2,3,4])
```

```python
Agewise_TB_Cases_For_LGA = block2c_2021_flattened.groupby(['LGA','Age_Gr'])['TB_Cases'].sum().reset_index()
```
