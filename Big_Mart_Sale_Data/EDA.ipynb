```python
item_weight = {}

items_with_weights = df[df.Item_Weight.notnull()]

for i,row in items_with_weights.iterrows():
    item_weight[row['Item_Identifier']] = row['Item_Weight']

item_weight
```

##### Outliers

```python
def calculate_outliers(data,col):
    col = sorted(data[col])
    Q1,Q3 = np.quantile(col,[0.25, 0.75])

    IQR = Q3-Q1

    lower_limit = Q1- (1.5*IQR)
    upper_limit = Q3+ (1.5*IQR)

    return lower_limit,upper_limit
```

```python
lower_limit,upper_limit = calculate_outliers(df,'Item_Visibility')

df['Item_Visibility'] = np.where(df['Item_Visibility']>upper_limit,upper_limit,df['Item_Visibility'])
```

```python
lower_limit,upper_limit = calculate_outliers(df,'Item_Outlet_Sales')

df['Item_Outlet_Sales'] = np.where(df['Item_Outlet_Sales']>upper_limit,upper_limit,df['Item_Outlet_Sales'])
```
